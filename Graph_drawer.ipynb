```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def import_data_for_heatmap():
    with open("phishing.data", "r") as f:
        data = f.read().splitlines()
    data = [list(map(int, line.split(","))) for line in data]

    full_data = np.array(data)

    sorted_data = full_data[full_data[:, -1].argsort()]

    X_sorted = sorted_data[:, :-1]
    y_sorted = sorted_data[:, -1]

    return X_sorted, y_sorted
```

```python
Sigmas=[0.1,0.5,1,5,10]
accuracies = [67.95,73.89,93.74,91.19,90.93]
plt.figure(figsize=(10,6))
sns.lineplot(x=Sigmas, y=accuracies, marker='o')
plt.xscale('log')
plt.xlabel('Sigma')
plt.ylabel('Accuracy (%)')
plt.title('SVM Accuracy vs Sigma')
plt.grid()
plt.show()
```

```python
X_vis, y_vis = import_data_for_heatmap()

plt.figure(figsize=(14, 8))


sns.heatmap(X_vis, cmap="coolwarm", cbar=False, yticklabels=False)


split_index = np.where(y_vis == 1)[0][0]


plt.axhline(y=split_index, color="black", linestyle="--", linewidth=2)

plt.text(
    X_vis.shape[1] + 1,
    split_index / 2,
    "Klasa -1 (Nie-Phishing)",
    va="center",
    color="blue",
    weight="bold",
)
plt.text(
    X_vis.shape[1] + 1,
    split_index + (len(y_vis) - split_index) / 2,
    "Klasa 1 (Phishing)",
    va="center",
    color="red",
    weight="bold",
)

plt.title("Mapa cech zbioru")
plt.xlabel("Indeks cechy (Kolumny)")
plt.ylabel("Obserwacje (Wiersze)")
plt.show()
```
